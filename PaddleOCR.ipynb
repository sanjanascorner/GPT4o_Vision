```python
from PIL import Image
import cv2
import numpy as np
import pandas as pd
import tensorflow as tf
import layoutparser as lp
from paddleocr import PaddleOCR, draw_ocr
from google.colab import files
from google.colab.patches import cv2_imshow
import io
```

```python
def upload_image():
    uploaded = files.upload()
    file_name = next(iter(uploaded))
    return file_name
```

```python
print("Please upload an image file.")
image_path = upload_image()
```

```python
image = cv2.imread(image_path)
image = image[..., ::-1]
```

```python

model = lp.Detectron2LayoutModel('lp://PubLayNet/faster_rcnn_R_50_FPN_3x/config',
                                 extra_config=["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.8],
                                 label_map={0: "Text", 1: "Title", 2: "List", 3:"Table", 4:"Figure"})
layout = model.detect(image) #detecting layout
lp.draw_box(image, layout, box_width=3)
```

```python
layout
```

```python
x_1, y_1, x_2, y_2 = 100,100, 1087, 864  # Default values
for l in layout:
    if l.type == 'Table':
        x_1 = int(l.block.x_1)
        y_1 = int(l.block.y_1)
        x_2 = int(l.block.x_2)
        y_2 = int(l.block.y_2)
        print(f"Table coordinates: {x_1}, {y_1}, {x_2}, {y_2}")
        break
```

```python
table_image = image[y_1:y_2, x_1:x_2]
cv2.imwrite('extracted_table.jpg', cv2.cvtColor(table_image, cv2.COLOR_RGB2BGR))
print("Performing OCR on the extracted table...")
ocr = PaddleOCR(lang='en')
output = ocr.ocr(table_image)
```

```python
boxes = [line[0] for line in output[0]]
texts = [line[1][0] for line in output[0]]
probabilities = [line[1][1] for line in output[0]]
```

```python
image_with_ocr = table_image.copy()
for box, text in zip(boxes, texts):
    cv2.rectangle(image_with_ocr, (int(box[0][0]), int(box[0][1])), (int(box[2][0]), int(box[2][1])), (0, 0, 255), 2)
    cv2.putText(image_with_ocr, text, (int(box[0][0]), int(box[0][1])), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (222,0, 0), 1)
```

```python
cv2_imshow(image_with_ocr)
```

```python
cv2.imwrite('ocr_result.jpg', cv2.cvtColor(image_with_ocr, cv2.COLOR_RGB2BGR))
files.download('ocr_result.jpg')
```

```python
ocr_text_content = '\n'.join(texts)
with open('ocr_results.txt', 'w') as f:
    f.write(ocr_text_content)
```
